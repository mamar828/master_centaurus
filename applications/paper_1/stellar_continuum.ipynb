```python
import numpy as np
import graphinglib as gl
from astropy.io import fits
```

```python
f = fits.open("data/loki.single_star_hot_z-1.0.fits")
spec = gl.Curve(np.array([*f[1].data]) * 1e-4, f[3].data)  # convert to microns
%matplotlib inline
gl.SmartFigure(elements=[spec], x_lim=(1.64, 3.14), y_lim=(0.25e-5, 1.5e-5)).show()
```

```python
arr = np.loadtxt("data/sed_lhs1140b_ord1.csv", delimiter=",", skiprows=1)
spec = gl.Curve(arr[:, 0], arr[:, 1])  # convert to microns
%matplotlib tk
gl.SmartFigure(elements=[spec]).show()
```
